```python
from env_utils import get_building_footprint, get_polygon_area, get_building_map
from pvwatts_api import get_pv_annual_kWh_per_kW, KW_PER_M2
```

```python
address = "AECOM, Calgary, AB"
pct_roof_area_available = 0.50
```

```python
footprint_geojson, lat, lon = get_building_footprint(address)
roof_area = get_polygon_area(footprint_geojson) if footprint_geojson else 0
available_roof_area = roof_area * pct_roof_area_available
```

```python
pv_system_size_kW = available_roof_area * KW_PER_M2
kWh_per_kW = get_pv_annual_kWh_per_kW(lat, lon)
pv_system_energy_kWh = pv_system_size_kW * kWh_per_kW

print(f"Estimated available roof area for PV: {available_roof_area:,.1f} m²")
print(f"Estimated PV system size: {pv_system_size_kW:,.1f} kW")
print(f"Estimated annual energy production: {round(pv_system_energy_kWh, -2):,.0f} kWh")
```
